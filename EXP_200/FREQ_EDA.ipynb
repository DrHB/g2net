```python


from sklearn.metrics import roc_auc_score
def get_score(df):
    return roc_auc_score(df['target'], df['pred'])

def get_freq(left, right, df):
    df_ = df.query(f"freq>{left} & freq<{right}")
    return df_

def get_snr(left, right, df):
    df_ = pd.concat([df.query(f"snr>{left} & snr<{right}"), df.query("snr==0")])
    return df_

def get_freqency(file):
    file = Path(file)
    with h5py.File(file, "r") as f:
        try:
            filename = file.stem
            freq_hz = np.mean(f["frequency_Hz"])
        except:
            filename = file.stem
            f = f[filename]
            freq_hz = np.mean(list(f["frequency_Hz"]))
        return freq_hz
```

```python
snr_left = 10
snr_right = 40
for i in range(10):
    print("________")
    df = pd.read_csv(
        f"../EXP_200/EXP_200_BASELINE_CASHE_V4/EXP_200_BASELINE_CASHE_V4_convnext_large_in22k_0_{i}_oof.csv"
    )
    df["freq"] = df["id"].apply(lambda x: get_freqency(x))
    print(
        "rho_10_40, freq  50  100:",
        "sample size:",
        get_snr(snr_left, snr_right, (get_freq(50, 100, df))).shape[0],
        get_score(get_snr(snr_left, snr_right, (get_freq(50, 100, df)))),
    )
    print(
        "rho_10_40, freq 100, 200:",
        "sample size:",
        get_snr(snr_left, snr_right, (get_freq(100, 200, df))).shape[0],
        get_score(get_snr(snr_left, snr_right, (get_freq(100, 200, df)))),
    )
    print(
        "rho_10_40, freq 200, 300:",
        "sample size:",
        get_snr(snr_left, snr_right, (get_freq(200, 300, df))).shape[0],
        get_score(get_snr(snr_left, snr_right, (get_freq(200, 300, df)))),
    )
    print(
        "rho_10_40, freq 300, 400:",
        "sample size:",
        get_snr(snr_left, snr_right, (get_freq(300, 400, df))).shape[0],
        get_score(get_snr(snr_left, snr_right, (get_freq(300, 400, df)))),
    )
    print(
        "rho_10_40, freq 400, 500:",
        "sample size:",
        get_snr(snr_left, snr_right, (get_freq(400, 500, df))).shape[0],
        get_score(get_snr(snr_left, snr_right, (get_freq(400, 500, df)))),
    )
```
